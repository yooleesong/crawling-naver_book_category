```python
import requests
```

```python
res = requests.get('https://book.naver.com/')
soup = BeautifulSoup(res.text)
attr = {
    "id": "left_category",
    "class": "category"
}
text = soup.find('div', attr = attr)
```

```python
print(text)
```

```python
span = soup.find('div', id="left_category").find_all('a')
ls = []
for i in span:
    ls.append(i.get_text())
ls
```

```python
main_cate = soup.find('div', id="left_category").find_all('span')
main_ls = []
for v in main_cate:
    main_ls.append(v.get_text())
main_ls
```

```python
sub_list = []
index = -1
for i in ls:
    if i in main_ls:
        sub_list.append([])
        index = index + 1
    else:
        sub_list[index].append(i)

sub_list
        
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(sub_list).transpose()
```

```python
df.columns = main_ls
df
```

```python
df =df.fillna(value='')
df
```
